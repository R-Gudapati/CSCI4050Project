```python
import pandas as pd
import numpy as np
import random
```

```python
books = pd.read_csv('books.csv', encoding='latin-1', on_bad_lines='skip',
                    sep=';')
books = books.sample(frac=0.00015, random_state=457)

books.drop(['Image-URL-S', 'Image-URL-M'], axis=1)
#print(books.columns)

final_books = pd.DataFrame(columns=['isbn',
    'title',
    'author',
    'edition',
    'category',
    'publisher',
    'publication_year',
    'image_url',
    'quantity',
    'selling_price',
    'buying_price',
    'min_threshold'])

final_books['title'] = books['Book-Title']
final_books['author'] = books['Book-Author']
final_books['publication_year'] = books['Year-Of-Publication']
final_books['publisher'] = books['Publisher']
final_books['image_url'] = books['Image-URL-L']
final_books['isbn'] = books['ISBN']

book_genres = [
    'Fiction',
    'Mystery/Thriller',
    'Romance',
    'Science Fiction',
    'Fantasy',
    'Horror',
    'Historical Fiction',
    'Adventure',
    'Dystopian',
    'Young Adult',
    "Children's",
    'Non-Fiction',
    'Biography/Autobiography',
    'Self-Help',
    'Health/Fitness',
    'Travel',
    'Science/Nature',
    'History',
    'True Crime',
    'Poetry',
    'Classic Literature',
    'Humor/Comedy',
    'Drama/Play',
    'Graphic Novel/Comics',
    'Anthology/Short Stories'
]

editions_list = ['1st', '2nd', '3rd', '4th', '5th']

numbers_list = list(range(10, 101))

final_books['category'] = random.choices(book_genres, k=len(final_books))
final_books['edition'] = random.choices(editions_list, k=len(final_books))
final_books['quantity'] = random.choices(numbers_list, k=len(final_books))

min_price = 9.99
max_price = 19.99
final_books['selling_price'] = np.random.uniform(min_price, max_price, len(final_books))
final_books['selling_price'] = final_books['selling_price'].round(2)

final_books['buying_price'] = np.random.uniform(min_price-5, max_price-5, len(final_books))
final_books['buying_price'] = final_books['buying_price'].round(2)

final_books['min_threshold'] = 8

final_books.to_csv('book_list.csv', index=False)
```
